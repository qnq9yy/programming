Now do it again, with slightly less direction:

8. Create a grid of 100 equally spaced points ranging from -6.5 to 6.5.

```python
import numpy as np
import matplotlib.pyplot as plt

print("Q2 parts 1 to 5")
x = np.linspace(0,1,50)
y = np.log(x)
z = np.exp(x)
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()

print("Q2 parts 6 and 7")
x = np.linspace(0,1,50)
y = np.log(x)
z = np.exp(x)
plt.scatter(x,y, label ='Natural Log')
plt.scatter(x,z, label='Exponential')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='lower right')
plt.title('Natural Log and Exponential Functions')
plt.show()

print("# Q2 parts 8 and 9")
x = np.linspace(-6.5,6.5,100)
y = np.sin(x)
z = np.cos(x)
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()

print("Q2 part 11")
x = np.linspace(-6.5,6.5,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, label ='Sine')
plt.plot(x,z, label='Cosine')
plt.show()

print("Q2 part 12")
x = np.linspace(-6.5,6.5,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, label ='Sine')
plt.plot(x,z, label='Cosine')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='lower left')
plt.title('Sine and Cosine')
plt.show()
```
